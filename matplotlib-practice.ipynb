### Naming X & Y AXIS 

```python
x = [1,2,5]
y = [5,6,7]

#Adding Title for Graph!
plt.title('Some Random Graph')

#Naming X-Axis!
plt.xlabel('X-AXIS')

#Naming Y-Axis!
plt.ylabel('Y-AXIS')

plt.plot(x,y)
plt.show()
```
