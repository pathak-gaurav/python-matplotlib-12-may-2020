### MATPLOTLIB TUTORIALS

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### BASIC GRAPH

```python
plt.plot([1,2,4],[4,5,6])
plt.show()
```

### Naming X & Y AXIS 

```python
x = [1,2,5]
y = [5,6,7]

#Adding Title for Graph!
plt.title('Some Random Graph')

#Naming X-Axis!
plt.xlabel('X-AXIS')

#Naming Y-Axis!
plt.ylabel('Y-AXIS')

plt.plot(x,y)
plt.show()
```

### Changing default font and size 

```python
x=[1,2,4]
y=[5,3,4]

# Adding Title and changing font size of Graph 
plt.title('Some Random Graph',fontdict={'fontname':'Comic Sans MS','fontsize':'20'})

#Adding Label and Changing font Size of X-Axis
plt.xlabel('X-Axis',fontdict={'fontname':'Comic Sans MS', 'fontsize':'10'})

# Adding Label and Changing font Size of Y-Axis 
plt.ylabel('Y-Axis',fontdict={'fontname':'Comic Sans MS','fontsize':'10'})

plt.plot(x,y)
plt.show()
```

### Adding tick on Axis

```python
x=[0,1,2,3,4]
y=[2,4,6,8,10]

plt.title('Some Random Graph',fontdict={'fontname':'Comic Sans MS','fontsize':'20'})
plt.xlabel('X-Axis',fontdict={'fontname':'Comic Sans MS','fontsize':'10'})
plt.ylabel('Y-Axis',fontdict={'fontname':'Comic Sans MS','fontsize':'10'})

plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,2,4,6,8,10])

#Legend i.e. Y is 2 times of X : 2X
plt.plot(x,y,label='2x',color='Maroon',linewidth=2,marker='.',markersize='14',markeredgecolor='blue',linestyle='--')
plt.legend()

plt.show()
```

```python
plt.plot(x,y,'rhe:',label='2X')
plt.legend()
```

### Graphs with numpy

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.title('Some Random Graph',fontdict={'fontname':'Comic Sans MS','fontsize':'16'})
plt.xlabel('X-Axis',fontdict={'fontname':'Comic Sans MS','fontsize':'10'})
plt.ylabel('Y-Axis',fontdict={'fontname':'Comic Sans MS','fontsize':10})

#plt.yticks([0,2,4,6,8,10])
plt.xticks([0,1,2,3,4])

#Linear Number 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5],x2[:5]**2,'rp:',label='x^2')
plt.plot(x2[4:],x2[4:]**2,'r')

plt.plot(x,y, 'mp--',label='2X')
plt.legend()
plt.show()
```
